```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.title("Centralizated Loss")
plt.xlabel("Rounds")
plt.ylabel("Loss")
y1 = pd.read_csv("//home/joaoneto/biometria/fl_behavior/scomnet/250_frames_1.csv", usecols=["loss"])
y2 = pd.read_csv("//home/joaoneto/biometria/fl_behavior/scomnet/500_frames_1.csv", usecols=["loss"])
y3 = pd.read_csv("//home/joaoneto/biometria/fl_behavior/scomnet/1000_frames_1.csv", usecols=["loss"])

plt.plot(y1, label="Frames=250")
plt.plot(y2, label="Frames=500")
plt.plot(y3, label="Frames=1000")

plt.legend()
plt.grid(True)
plt.savefig("/home/joaoneto/Results/scomnet/plot/Centralizated_Loss.jpg")
```
